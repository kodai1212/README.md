```python
import aspose.cad as cad
#dwg/dxf変換
inputDWG = "drive/MyDrive/1.dwg";
outputDXF = "drive/MyDrive/out.dxf";

# Load DWG file
image = cad.Image.load(inputDWG)

# Specify DXF options
options = cad.imageoptions.DxfOptions()

# Save output DXF file
image.save(outputDXF, options);
```

```python
import ezdxf

def read_dxf(file_path):
    doc = ezdxf.readfile(file_path)
    modelspace = doc.modelspace()

    entities = []
    for entity in modelspace:
        entities.append(entity)

    return entities

file_path = 'drive/MyDrive/out.dxf'
entities = read_dxf(file_path)

for entity in entities[:10]:
    print(entity)
```

```python
import ezdxf

def read_dxf(file_path):
    doc = dxf
    modelspace = doc.modelspace()

    entities = []
    for entity in modelspace:
        if entity.dxftype() == 'POLYLINE':
            polyline_coords = [(vertex.dxf.location.x, vertex.dxf.location.y) for vertex in entity.vertices]
            entities.append(polyline_coords)

    return entities

file_path = 'drive/MyDrive/out.dxf'
entities = read_dxf(file_path)

for polyline_coords in entities:
    print("POLYLINE Coordinates:")
    for x, y in polyline_coords:
        print(f"({x}, {y})")
    print()
```
